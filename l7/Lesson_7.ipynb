```python
def transcribe_audio_to_text(audio_path):
    # Convert MP3 to WAV
    if audio_path.lower().endswith(".mp3"):
        audio = AudioSegment.from_mp3(audio_path)
        audio.export("/tmp/temp.wav", format="wav")
        audio_path = "/tmp/temp.wav"

    recognizer = sr.Recognizer()
    audio_file = sr.AudioFile(audio_path)
    with audio_file as source:
        audio_data = recognizer.record(source)
    return recognizer.recognize_google(audio_data)

# Example usage
audio_path = "/content/audio2.mp3"  # Replace with the path to your audio file
transcribed_text = transcribe_audio_to_text(audio_path)
print(transcribed_text)
```

```python
def ask_question(text, question):
    response = llm.complete(f" Based on the following text: {extracted_text}, answer the question: {transcribed_text}. give answer within 50 words, in plain english.")
    return response.text

# Example usage
question = transcribed_text
answer = ask_question(transcribed_text, question)
print(f"Question: {transcribed_text}")
print(f"Answer: {answer}")
```

```python
def text_to_speech(text, output_audio_path):
    tts = gTTS(text=text, lang='en')
    tts.save(output_audio_path)
    return output_audio_path

# Example usage
response_audio_path = text_to_speech(answer, "/content/response_audio.mp3")
ipd.display(ipd.Audio(response_audio_path))
```
